```
# Program Sederhana Menghitung Gaji Karyawan
# Menggunakan Operator, IF, dan Array

def hitung_gaji_karyawan():
    """
    Fungsi utama untuk menghitung gaji karyawan berdasarkan golongan dan jam lembur.
    """

    # Array Gaji Pokok
    # Index 0: Golongan A
    # Index 1: Golongan B
    # Index 2: Golongan C
    gaji_pokok_arr = [5000000, 6500000, 9500000]

    # Array Persen Lembur
    # Index 0: 1 Jam lembur (30%)
    # Index 1: 2 Jam lembur (32%)
    # Index 2: 3 Jam lembur (34%)
    # Index 3: 4 Jam lembur (36%)
    # Index 4: >= 5 Jam lembur (38%)
    persen_lembur_arr = [0.30, 0.32, 0.34, 0.36, 0.38]

    # --- Input dari Pengguna ---
    print("---------------------------------------------")
    print("PROGRAM PENGHITUNGAN GAJI KARYAWAN SEDERHANA")
    print("---------------------------------------------")

    # Input Golongan Karyawan dengan validasi
    while True:
        golongan_input = input("Masukkan Golongan Karyawan (A/B/C): ").strip().upper()
        if golongan_input in ["A", "B", "C"]:
            break
        else:
            print("Pilihan golongan tidak valid. Harap masukkan A, B, atau C.")

    # Input Jam Lembur dengan validasi
    while True:
        try:
            jam_lembur_input = int(input("Masukkan Jumlah Jam Lembur (angka, 0 jika tidak ada): "))
            if jam_lembur_input >= 0:
                break
            else:
                print("Jumlah jam lembur tidak boleh negatif.")
        except ValueError:
            print("Input jam lembur tidak valid. Harap masukkan angka.")

    # --- Proses Perhitungan ---

    # Menentukan Gaji Pokok berdasarkan Golongan
    gaji_pokok_karyawan = 0
    index_gaji = -1 # Untuk mengambil persentase lembur yang sesuai

    if golongan_input == "A":
        gaji_pokok_karyawan = gaji_pokok_arr[0]
        index_gaji = 0
    elif golongan_input == "B":
        gaji_pokok_karyawan = gaji_pokok_arr[1]
        index_gaji = 1
    elif golongan_input == "C":
        gaji_pokok_karyawan = gaji_pokok_arr[2]
        index_gaji = 2

    # Menghitung Gaji Lembur
    gaji_lembur_karyawan = 0
    persentase_lembur_digunakan = 0

    if jam_lembur_input > 0:
        if jam_lembur_input == 1:
            persentase_lembur_digunakan = persen_lembur_arr[0]
        elif jam_lembur_input == 2:
            persentase_lembur_digunakan = persen_lembur_arr[1]
        elif jam_lembur_input == 3:
            persentase_lembur_digunakan = persen_lembur_arr[2]
        elif jam_lembur_input == 4:
            persentase_lembur_digunakan = persen_lembur_arr[3]
        elif jam_lembur_input >= 5:
            persentase_lembur_digunakan = persen_lembur_arr[4]

        # Gaji lembur dihitung dari Gaji Pokok sesuai golongan
        gaji_lembur_karyawan = gaji_pokok_karyawan * persentase_lembur_digunakan

    # Menghitung Jumlah Penghasilan Total
    jumlah_penghasilan = gaji_pokok_karyawan + gaji_lembur_karyawan

    # --- Output Hasil Perhitungan ---
    print("\n--- Rincian Penghasilan Karyawan ---")
    print(f"Golongan Karyawan : {golongan_input}")
    print(f"Gaji Pokok        : Rp. {gaji_pokok_karyawan:,.0f}".replace(",", "."))
    print(f"Jam Lembur        : {jam_lembur_input} Jam")
    if jam_lembur_input > 0:
        print(f"Persentase Lembur : {persentase_lembur_digunakan*100:.0f}%")
    print(f"Gaji Lembur       : Rp. {gaji_lembur_karyawan:,.0f}".replace(",", "."))
    print("--------------------------------------")
    print(f"Total Penghasilan : Rp. {jumlah_penghasilan:,.0f}".replace(",", "."))
    print("--------------------------------------\n")

# Menjalankan fungsi utama program
if __name__ == "__main__":
    hitung_gaji_karyawan()
```
